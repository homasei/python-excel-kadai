```python
import openpyxl
import time

#ワークブックの準備
workbook = openpyxl.Workbook()
ws = workbook.active

#日付の取得
today1 = time.strftime("%Y/%m/%d")
today2 = time.strftime("%Y%m%d")

#請求先情報
post_code = "〒101-0022"
address = "東京都千代田区神田練堀町300"
company = "株式会社ABC"
telephone_number = "TEL:03-1234-5678"
fax_number = "03-1234-5678"
charge_name = "鈴木一郎"

#請求書発行番号
issue = "0001"

#請求書フォーマット
ws["B2"] = "請求書"
ws["B4"] = company
ws["B5"] = post_code +" "+ address
ws["B6"] = telephone_number +" "+ fax_number
ws["B7"] = "担当者名:" + charge_name + "様"

ws["F4"] = "No."
ws["F5"] = "日付"
ws["G4"] = issue
ws["G5"] = today1

ws.cell(row=10, column=2, value="商品名")
ws.cell(row=10, column=3, value="数量")
ws.cell(row=10, column=4, value="単価")
ws.cell(row=10, column=5, value="金額")

ws["B15"] = "小計"
ws["B16"] = "消費税"
ws["B17"] = "合計"

#請求データ
data = [
    ["商品A", 2, 10000, 20000],
    ["商品B", 1, 15000, 15000], 
]

start_row = 11
start_column = 2

for i, row in enumerate(data):
    for j, value in enumerate(row):
        cell = ws.cell(row = start_row + i,column = start_column + j)
        cell.value = value

#合計額の算出
x = 0
for i in data:
    x += i[3]


#消費税額の算定

tax_rate = 0.1
tax = x * tax_rate
ws["E15"] = x
ws["E16"] = round(tax)
ws["E17"] = x + round(tax)


        
#データの保存
workbook.save("請求書_"+ today2 +".xlsx")
```
